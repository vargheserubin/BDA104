# Loading Graphs in NetworkX

 Using the following Data sets regarding to the airports and their flight information
- create a directed graph based on flight information
- using pageRank algorithm (avaialable in networkx) in python, find the name of the first 10 important airports


```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#airports
airport_col = ['ID', 'Name', 'City', 'Country','IATA', 'ICAO', 'Lat', 'Long', 'Alt', 'Timezone', 'DST', 'Tz database time zone', 'type', 'source']
airport_df = pd.read_csv("https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat",names = airport_col, index_col = 0)

route_cols = ['Airline', 'Airline ID', 'Source Airport', 'Source Airport ID','Dest Airport', 'Dest Airport ID', 'Codeshare', 'Stops', 'equipment']
routes_df = pd.read_csv("https://raw.githubusercontent.com/jpatokal/openflights/master/data/routes.dat", names = route_cols)
```

```python
airport_df.head(5)
```

```python
routes_df.head(10)
```

```python
import networkx as nx
network = nx.Graph()
```

```python
for airport in routes_df:
    network.add_node(airport[0], id=airport[0], name=airport[1], city=airport[2], country=airport[3], iata=airport[4], 
                     icao=airport[5],
                     lat=airport[6], 
                     long=airport[7], alt=airport[8], offset=airport[9], 
                     daylight=airport[10], timezone=airport[11], type=airport[12], source=airport[13])
```
